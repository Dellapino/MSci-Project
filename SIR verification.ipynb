```python
import networkx as nx
import network_stats as ns

degree = 35
size = 500
degree_dist = []
for i in range(size):
    degree_dist.append(degree)
    
G = nx.configuration_model(degree_dist, create_using = nx.Graph)

def initialise_weight(G, weight):
    nodes = list(G.nodes)
    for node in nodes:
        connections = list(G.neighbors(node))
        for connection in connections:
            G[node][connection]['weight'] = weight
    return G

G = initialise_weight(G, 1)
node_measures = ns.measure_nodes(G)
network_measures = ns.measure_network(G)
def print_dict(dict):
    for key in list(dict.keys()):
        print(key + ' : ' + str(dict[key]))
    print('\n')
        
print('Node measures')
print_dict(node_measures)
print('Network measures')
print_dict(network_measures)

'''

can make network of arbitrary size with uniform mixing

to compare will need to remove weights from network (set = 1)

how is this different to comparing with ODE directly?

already have basic SIR comparison

aim is to provide conditions in which SIR is valid and show IF model works in those
conditions (uniform weight and mixing)

then show that SIR model breaks down and IF can work for weighted networks.

'''
```
