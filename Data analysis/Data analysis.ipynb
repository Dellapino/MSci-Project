```python
# -*- coding: utf-8 -*-
"""
Created on Wed Jan 18 18:48:57 2023

@author: MSI
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

single = pd.read_csv('single_slit.csv')

double = pd.read_csv('double_slit.csv')
#print(single)
#print(double)

xs1 = np.array(single.iloc[:,[1]]).flatten()
xs2 = np.array(single.iloc[:,[2]]).flatten()
ms = np.array(single.iloc[:,[0]]).flatten()

xd = np.array(double.iloc[:,[1]]).flatten()
md = np.array(double.iloc[:,[0]]).flatten()


xd = np.array(double['x'])
md = np.array(double['m'])

xd = xd.astype(float)
md = md.astype(float)
```

```python
def line(x, m, n):
    return (x*m)+n 

guess_m = 1
guess_n = 1

p0 = [guess_m, guess_n]
```

```python
lambd = 670*10**(-6)
f = 500

errors = np.empty(5)
errors.fill(1)


fit = curve_fit(line, ms, xs1, p0 = p0)

data_fit = line(ms,*fit[0])


plt.plot(ms,xs1,'x')
plt.plot(ms, data_fit)
plt.xlabel('m [-]')
plt.ylabel('x [mm]')
plt.errorbar(ms,xs1,yerr=errors, fmt='o', mew=1, ms=1, capsize=4)
plt.show()

grad = fit[0][0]
grad_unc = np.sqrt(fit[1][0][0])

a = lambd*f/grad
a_unc = a* grad_unc/grad

print('The width of the slit is ' + str(a) + '+/-' + str(a_unc) + ' mm')
```

```python
#%% Set 2

lambd = 670*10**(-6)
f = 500

errors = np.empty(5)
errors.fill(1)

fit = curve_fit(line, ms, xs2, p0 = p0, sigma = 1/errors)

data_fit = line(ms,*fit[0])


plt.plot(ms,xs2,'x')
plt.plot(ms, data_fit)
plt.xlabel('m [-]')
plt.ylabel('x [mm]')
plt.errorbar(ms,xs2,yerr=errors, fmt='o', mew=1, ms=1, capsize=4)
plt.show()

grad = fit[0][0]
grad_unc = np.sqrt(fit[1][0][0])

a2 = lambd*f/grad
a2_unc = a* grad_unc/grad

print('The width of the slit is ' + str(a2) + '+/-' + str(a2_unc) + ' mm')
```
